```python
import numpy as np # mengelola data numerik
import pandas as pd # mengelola data excel
import matplotlib.pyplot as plt # untuk membuat grafik
```

```python
# Impor Dataset
dataset = pd.read_csv('kesehatan_umum.csv')
```

```python
dataset.head()
```

```python
x = dataset.iloc[:, [2, 3]].values               # membagi nilai x
y = dataset.iloc[:, -1].values                  # membagi nilai y
```

```python
print (x)
```

```python
print (y)
```

```python
len (y)
```

# **Split Dataset**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print (x_train)
```

```python
print (x_test)
```

```python
print (y_train)
```

```python
print (y_test)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print (x_train)
```

```python
print (x_test)
```

# **Latih Model KNN Dengan Data Training**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(x_train, y_train)
```

# **Memprediksi Hasil Set Pengujian**

```python
y_pred = classifier.predict(x_test)
```

# **Membuat Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print (cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test, y_pred)
print ("Tingkat Akurasi : %d persen " % (akurasi*100))
```

# **Visualisasi Data Training dan Testing**

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                    np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan Naive Bayes (Data Training)')
plt.xlabel('Tekanan Darah Saat Istirahat')
plt.ylabel('Kolesterol')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                    np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan Naive Bayes (Data Testing)')
plt.xlabel('Tekanan Darah Saat Istirahat')
plt.ylabel('Kolesterol')
plt.legend()
plt.show()
```
